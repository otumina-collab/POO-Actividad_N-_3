```python
# Actividad N°3 ejercicio 8.2
#Solucion
import math
import ipywidgets as widgets
from IPython.display import display

# Mantiene la lógica original para el cálculo de notas.

class Notas:
    """
    Clase que define la lista de notas numéricas y
    realiza las operaciones estadísticas.
    """
    def __init__(self):
# Lista para almacenar las 5 notas
        self.listaNotas = [0.0] * 5

    def calcularPromedio(self):
        """Calcula el promedio de las notas."""

# Se asegura de que la lista no esté vacía antes de calcular
        if not self.listaNotas:
            return 0.0
        suma = sum(self.listaNotas)
        return suma / len(self.listaNotas)

    def calcularDesviacion(self):
        """Calcula la desviación estándar."""

# Se asegura de que la lista tenga elementos
        if not self.listaNotas or len(self.listaNotas) == 0:
            return 0.0

        prom = self.calcularPromedio()
        suma_cuadrados = 0

# Fórmula: Sumatoria de (nota[i] - promedio)^2
        for nota in self.listaNotas:
            suma_cuadrados += math.pow(nota - prom, 2)

# Raíz cuadrada de la varianza
# Nota: Usamos la fórmula para la desviación estándar poblacional
        return math.sqrt(suma_cuadrados / len(self.listaNotas))

    def calcularMenor(self):
        """Encuentra el valor menor."""
        return min(self.listaNotas) if self.listaNotas else 0.0

    def calcularMayor(self):
        """Encuentra el valor mayor."""
        return max(self.listaNotas) if self.listaNotas else 0.0

# INTERFAZ GRÁFICA (IPYWIDGETS)
# Implementa la interfaz de usuario con widgets.

# Instancia de la clase de cálculos
calculadora_notas = Notas()

# Crear widgets de entrada para las 5 notas
input_notas = [widgets.FloatText(description=f"Nota {i+1}:", value=0.0) for i in range(5)]

# Crear widgets de salida para los resultados
output_promedio = widgets.Label(value="Promedio: --")
output_desviacion = widgets.Label(value="Desviación Estándar: --")
output_mayor = widgets.Label(value="Nota Mayor: --")
output_menor = widgets.Label(value="Nota Menor: --")

# Crear botones
boton_calcular = widgets.Button(description="Calcular")
boton_limpiar = widgets.Button(description="Limpiar")

# Función para calcular las notas y mostrar resultados
def calcular_ipywidgets(b):
    notas_ingresadas = []
    error = False
    for i, campo in enumerate(input_notas):
        try:
            nota = campo.value
            if nota < 0 or nota > 100:
                 raise ValueError("La nota debe estar en un rango válido (0-100).")
            notas_ingresadas.append(nota)
        except ValueError as e:
            print(f"Error en Nota {i+1}: {e}")
            error = True
            break

    if not error:
        calculadora_notas.listaNotas = notas_ingresadas
        try:
            promedio = calculadora_notas.calcularPromedio()
            desviacion = calculadora_notas.calcularDesviacion()
            mayor = calculadora_notas.calcularMayor()
            menor = calculadora_notas.calcularMenor()

            output_promedio.value = f"Promedio: {promedio:.2f}"
            output_desviacion.value = f"Desviación Estándar: {desviacion:.2f}"
            output_mayor.value = f"Nota Mayor: {mayor:.1f}"
            output_menor.value = f"Nota Menor: {menor:.1f}"
        except Exception as e:
            output_promedio.value = f"Error de cálculo: {e}"
            output_desviacion.value = ""
            output_mayor.value = ""
            output_menor.value = ""

# Función para limpiar los campos
def limpiar_ipywidgets(b):
    for campo in input_notas:
        campo.value = 0.0
    output_promedio.value = "Promedio: --"
    output_desviacion.value = "Desviación Estándar: --"
    output_mayor.value = "Nota Mayor: --"
    output_menor.value = "Nota Menor: --"

# Asignar las funciones a los botones
boton_calcular.on_click(calcular_ipywidgets)
boton_limpiar.on_click(limpiar_ipywidgets)

# Organizar los widgets en un contenedor vertical
interfaz_calculadora = widgets.VBox(
    [
        widgets.Label(value="Notas:", style={'font_weight': 'bold'}),
        *input_notas,
        widgets.HBox([boton_calcular, boton_limpiar]),
        widgets.Label(value="Resultados:", style={'font_weight': 'bold'}),
        output_promedio,
        output_desviacion,
        output_mayor,
        output_menor,
    ]
)

# Mostrar la interfaz
display(interfaz_calculadora)
```

```python
# Importar módulos de ipywidgets y Python necesarios
# ipywidgets se usan para crear la interfaz de usuario interactiva
from ipywidgets import VBox, HBox, Label, FloatText, Dropdown, Button, Output, RadioButtons
# IPython.display se usa para renderizar los widgets en la salida del notebook
from IPython.display import display
# El módulo math es necesario para operaciones matemáticas (como pi y sqrt)
import math

# CLASE 1: LÓGICA DE CÁLCULO DE FIGURAS GEOMÉTRICAS
# Implementa las fórmulas matemáticas para las tres figuras.
class Geometria:
    """
    Clase que centraliza las operaciones matemáticas para Cilindro, Esfera y Pirámide.
    Contiene métodos para calcular el volumen y la superficie de cada figura.
    Incluye métodos para cálculos de la pirámide basados tanto en la altura como en la apotema inclinada.
    """
    def __init__(self):

# Inicializar el valor de Pi
        self.PI = math.pi
# CILINDRO
    def calcular_volumen_cilindro(self, radio, altura):
        """
        Calcula el volumen de un cilindro.
        Fórmula: V = pi * r^2 * h
        Args:
            radio (float): El radio de la base del cilindro.
            altura (float): La altura del cilindro.
        Returns:
            float: El volumen del cilindro.
        """
        return self.PI * (radio ** 2) * altura

    def calcular_superficie_cilindro(self, radio, altura):
        """
        Calcula la superficie total de un cilindro.
        Fórmula: A = 2 * pi * r * h + 2 * pi * r^2
        Args:
            radio (float): El radio de la base del cilindro.
            altura (float): La altura del cilindro.
        Returns:
            float: La superficie total del cilindro.
        """
        area_lateral = 2 * self.PI * radio * altura
        area_bases = 2 * self.PI * (radio ** 2)
        return area_lateral + area_bases

# ESFERA
    def calcular_volumen_esfera(self, radio):
        """
        Calcula el volumen de una esfera.
        Fórmula: V = (4/3) * pi * r^3
        Args:
            radio (float): El radio de la esfera.
        Returns:
            float: El volumen de la esfera.
        """
        return (4/3) * self.PI * (radio ** 3)

    def calcular_superficie_esfera(self, radio):
        """
        Calcula la superficie de una esfera.
        Fórmula: A = 4 * pi * r^2
        Args:
            radio (float): El radio de la esfera.
        Returns:
            float: La superficie de la esfera.
        """
        return 4 * self.PI * (radio ** 2)

# PIRÁMIDE (Base Cuadrada)
    def calcular_volumen_piramide(self, lado_base, apotema_inclinada):
        """
        Calcula el volumen de una pirámide de base cuadrada utilizando el lado de la base y la apotema inclinada.
        Calcula la altura (h) a partir del lado de la base (L) y la apotema inclinada (hs).
        Fórmula: V = (1/3) * L^2 * h
        h = sqrt(hs^2 - (L/2)^2)
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            apotema_inclinada (float): La apotema de una de las caras laterales (distancia del centro del lado de la base al vértice superior).
        Returns:
            float: El volumen de la pirámide.
        Raises:
            ValueError: Si apotema_inclinada es menor o igual a lado_base / 2, resultando en un número no positivo bajo la raíz cuadrada.
        """
        apotema_base = lado_base / 2

# Calcula la altura usando el teorema de Pitágoras
        if apotema_inclinada <= apotema_base:
             raise ValueError("La Apotema Inclinada debe ser estrictamente mayor que la mitad del Lado Base para formar una pirámide válida con altura positiva.")

        altura = math.sqrt(apotema_inclinada**2 - apotema_base**2)
        area_base = lado_base ** 2
        volumen = (1/3) * area_base * altura
        return volumen

    def calcular_superficie_piramide(self, lado_base, apotema_inclinada):
        """
        Calcula la superficie total de una pirámide de base cuadrada utilizando el lado de la base y la apotema inclinada.
        Fórmula: A = L^2 (área base) + 2 * L * hs (área lateral, suma de 4 triángulos)
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            apotema_inclinada (float): La apotema de una de las caras laterales.
        Returns:
            float: La superficie total de la pirámide.
        """
        area_base = lado_base ** 2

        perimetro_base = 4 * lado_base
        area_lateral = 0.5 * perimetro_base * apotema_inclinada

        return area_base + area_lateral

# PIRÁMIDE (Base Cuadrada)
    def calcular_volumen_piramide_por_altura(self, lado_base, altura):
        """
        Calcula el volumen de una pirámide de base cuadrada utilizando el lado de la base y la altura.
        Fórmula: V = (1/3) * L^2 * h
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            altura (float): La altura de la pirámide (distancia perpendicular del centro de la base al vértice superior).
        Returns:
            float: El volumen de la pirámide.
        """
        area_base = lado_base ** 2
        return (1/3) * area_base * altura

    def calcular_superficie_piramide_por_altura(self, lado_base, altura):
        """
        Calcula la superficie total de una pirámide de base cuadrada utilizando el lado de la base y la altura.
        Deriva la apotema inclinada (hs) de la altura para calcular el área lateral.
        Fórmula: A = L^2 + 2 * L * hs
        hs = sqrt(h^2 + (L/2)^2)
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            altura (float): La altura de la pirámide.
        Returns:
            float: La superficie total de la pirámide.
        Raises:
            ValueError: Si lado_base o altura son no positivos (ya manejado por la validación de entrada, pero es una buena práctica).
        """
        apotema_base = lado_base / 2

        apotema_inclinada = math.sqrt(altura**2 + apotema_base**2)

        area_base = lado_base ** 2

        perimetro_base = 4 * lado_base
        area_lateral = 0.5 * perimetro_base * apotema_inclinada

        return area_base + area_lateral

    def calcular_altura_piramide(self, lado_base, apotema_inclinada):
        """
        Calcula la altura de una pirámide de base cuadrada a partir del lado de la base y la apotema inclinada.
        Fórmula: h = sqrt(hs^2 - (L/2)^2)
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            apotema_inclinada (float): La apotema de una de las caras laterales.
        Returns:
            float: La altura de la pirámide.
        Raises:
            ValueError: Si apotema_inclinada es menor o igual a lado_base / 2.
        """
        apotema_base = lado_base / 2
        if apotema_inclinada <= apotema_base:
             raise ValueError("La Apotema Inclinada debe ser estrictamente mayor que la mitad del Lado Base para formar una pirámide válida con altura positiva.")
        return math.sqrt(apotema_inclinada**2 - apotema_base**2)

    def calcular_apotema_inclinada_piramide(self, lado_base, altura):
        """
        Calcula la apotema inclinada de una pirámide de base cuadrada a partir del lado de la base y la altura.
        Fórmula: hs = sqrt(h^2 + (L/2)^2)
        Args:
            lado_base (float): La longitud de un lado de la base cuadrada.
            altura (float): La altura de la pirámide.
        Returns:
            float: La apotema inclinada de la pirámide.
        """
        apotema_base = lado_base / 2
        return math.sqrt(altura**2 + apotema_base**2)


# WIDGETS DE LA INTERFAZ GRÁFICA (IPYWIDGETS)
# Define y organiza los widgets para la interacción del usuario.
# Selector de figura (Dropdown)
selector_figura = Dropdown(
    options=["Cilindro", "Esfera", "Pirámide"],
    value="Cilindro",
    description="Figura:"
)

# Widgets de entrada (FloatText) para dimensiones
widgets_entrada = {
    "radio": FloatText(description="Radio (cms):"),
    "altura": FloatText(description="Altura (cms):"),
    "lado_base": FloatText(description="Lado Base (cms):", layout={'display': 'none'}),
    "apotema_inclinada": FloatText(description="Apotema Inclinada (cms):", layout={'display': 'none'}),
    "altura_piramide_input": FloatText(description="Altura (cms):", layout={'display': 'none'})
}

# Widgets de salida (Label) para mostrar resultados
widgets_salida = {
    "volumen": Label(value="Volumen (cm³): --"),
    "superficie": Label(value="Superficie (cm²): --"),
}

# Botón para disparar el cálculo
boton_calcular = Button(description="Calcular")

# Botón para limpiar los campos
boton_limpiar = Button(description="Limpiar")

# Área de salida para mostrar mensajes (errores, advertencias, etc.)
salida_mensajes = Output()

# LÓGICA DE INTERACCIÓN Y VALIDACIÓN
# Conecta los widgets a la lógica de cálculo y maneja eventos.
def actualizar_visibilidad_entrada(figura_seleccionada):
    """
    Actualiza la visibilidad de los widgets de entrada en función de la figura seleccionada.
    Oculta las entradas no requeridas para la figura actual y muestra las que sí lo son.
    También borra los valores de las entradas ocultas.

    Args:
        figura_seleccionada (str): El nombre de la figura actualmente seleccionada.
    """
# Definir qué entradas son potencialmente requeridas para cada figura
    todas_entradas_posibles = ["radio", "altura", "lado_base", "apotema_inclinada", "altura_piramide_input"]

# Ocultar todas las entradas por defecto
    for key in todas_entradas_posibles:
        if key in widgets_entrada:
            widgets_entrada[key].layout.display = 'none'
            widgets_entrada[key].value = 0.0 # Borrar valores al ocultar

# Ahora, mostrar las entradas requeridas según la figura seleccionada
    if figura_seleccionada == "Cilindro":
        widgets_entrada["radio"].layout.display = 'flex'
        widgets_entrada["altura"].layout.display = 'flex'
    elif figura_seleccionada == "Esfera":
        widgets_entrada["radio"].layout.display = 'flex'
    elif figura_seleccionada == "Pirámide":
        # Mostrar siempre los tres inputs para Piramide según lo solicitado
        widgets_entrada["lado_base"].layout.display = 'flex'
        widgets_entrada["altura_piramide_input"].layout.display = 'flex'
        widgets_entrada["apotema_inclinada"].layout.display = 'flex'


def al_cambiar_figura(change):
    """
    Manejador de eventos para cuando la figura seleccionada cambia en el desplegable.
    Llama a actualizar_visibilidad_entrada y borra resultados y mensajes anteriores.

    Args:
        change (dict): Un diccionario que contiene información sobre el evento de cambio.
                       La clave 'new' contiene el nuevo valor del desplegable.
    """
    nueva_figura = change['new']
# Actualizar qué campos de entrada son visibles en función de la nueva figura
    actualizar_visibilidad_entrada(nueva_figura)
# Borrar los resultados mostrados en las etiquetas de salida
    widgets_salida["volumen"].value = "Volumen (cm³): --"
    widgets_salida["superficie"].value = "Superficie (cm²): --"
# Borrar cualquier mensaje anterior en el área de salida de mensajes
    with salida_mensajes:
        salida_mensajes.clear_output()


def validar_y_obtener_valor(key, nombre_campo):
    """
    Valida que el valor ingresado en un widget de entrada específico sea un número positivo.
    Imprime un mensaje de error en salida_mensajes si la validación falla.
    Solo valida si el widget está actualmente visible y tiene un valor mayor que 0.

    Args:
        key (str): La clave del widget de entrada en el diccionario widgets_entrada.
        nombre_campo (str): Un nombre amigable para el campo de entrada (usado en mensajes de error).

    Returns:
        float o None: El valor float positivo validado, o None si la validación falla o el widget no está visible.
    """
# Verificar si el widget existe y está actualmente visible antes de validar
    if key not in widgets_entrada or widgets_entrada[key].layout.display == 'none':
        return None

    widget = widgets_entrada[key]
    valor = widget.value
# Verificar si el valor es no positivo
    if valor <= 0:
# No imprimir error aquí, ya que el usuario podría dejar un campo en 0 para que se calcule.
# La lógica de validación más completa se hace en al_hacer_click_boton.
        return None # Retornar None si el valor no es positivo, pero no imprimir error todavía.

    return valor


def al_hacer_click_boton(b):
    """
    Manejador de eventos para cuando se hace clic en el botón 'Calcular'.
    Recupera las entradas, las valida, realiza los cálculos utilizando la clase Geometria,
    y actualiza los widgets de salida con los resultados o imprime errores en salida_mensajes.
    Para la Pirámide, intenta calcular el valor faltante si solo se proporcionan dos.

    Args:
        b (Button): El widget de botón que fue clickeado (no se usa directamente en la lógica de la función).
    """
# Borrar cualquier mensaje anterior del área de salida antes de mostrar nuevos
    with salida_mensajes:
        salida_mensajes.clear_output()

# Obtener la figura actualmente seleccionada
    figura = selector_figura.value
# Instanciar la clase Geometria para usar sus métodos de cálculo
    geometria = Geometria()

# Borrar resultados anteriores en las etiquetas de salida
    widgets_salida["volumen"].value = "Volumen (cm³): --"
    widgets_salida["superficie"].value = "Superficie (cm²): --"

    try:
# Lógica para Cilindro
        if figura == "Cilindro":
            radio = validar_y_obtener_valor("radio", "Radio")
            altura = validar_y_obtener_valor("altura", "Altura")

# Proceder con el cálculo solo si ambas entradas son válidas
            if radio is not None and altura is not None:
# Calcular volumen y superficie
                volumen = geometria.calcular_volumen_cilindro(radio, altura)
                superficie = geometria.calcular_superficie_cilindro(radio, altura)
# Actualizar etiquetas de salida con resultados formateados
                widgets_salida["volumen"].value = f"Volumen (cm³): {volumen:.2f}"
                widgets_salida["superficie"].value = f"Superficie (cm²): {superficie:.2f}"

# Lógica para Esfera
        elif figura == "Esfera":
# Validar y obtener el valor del radio
            radio = validar_y_obtener_valor("radio", "Radio")

# Proceder con el cálculo solo si la entrada es válida
            if radio is not None:
# Calcular volumen y superficie
                volumen = geometria.calcular_volumen_esfera(radio)
                superficie = geometria.calcular_superficie_esfera(radio)
# Actualizar etiquetas de salida con resultados formateados
                widgets_salida["volumen"].value = f"Volumen (cm³): {volumen:.2f}"
                widgets_salida["superficie"].value = f"Superficie (cm²): {superficie:.2f}"

# Lógica para Pirámide (Base Cuadrada) - Ahora siempre mostrando los tres inputs
        elif figura == "Pirámide":
            lado_base = validar_y_obtener_valor("lado_base", "Lado Base")
            altura_input = widgets_entrada["altura_piramide_input"].value
            apotema_inclinada_input = widgets_entrada["apotema_inclinada"].value

            altura_valida = altura_input > 0
            apotema_inclinada_valida = apotema_inclinada_input > 0

            volumen = None
            superficie = None
            altura_calculada = None
            apotema_inclinada_calculada = None

            if lado_base is not None:
                if altura_valida and apotema_inclinada_valida:
# Ambos proporcionados. Priorizar cálculo por altura.
                    altura_calculada = altura_input
                    volumen = geometria.calcular_volumen_piramide_por_altura(lado_base, altura_calculada)
                    superficie = geometria.calcular_superficie_piramide_por_altura(lado_base, altura_calculada)

                elif altura_valida:
# Solo altura proporcionada (o apotema_inclinada inválida/no proporcionada)
                    altura_calculada = altura_input
                    apotema_inclinada_calculada = geometria.calcular_apotema_inclinada_piramide(lado_base, altura_calculada)
                    widgets_entrada["apotema_inclinada"].value = round(apotema_inclinada_calculada, 2) # Mostrar el valor calculado redondeado
                    volumen = geometria.calcular_volumen_piramide_por_altura(lado_base, altura_calculada)
                    superficie = geometria.calcular_superficie_piramide_por_altura(lado_base, altura_calculada)

                elif apotema_inclinada_valida:
# Solo apotema inclinada proporcionada (o altura inválida/no proporcionada)
# Validar restricción geométrica para apotema_inclinada
                    if apotema_inclinada_input <= (lado_base / 2):
                        with salida_mensajes:
                             print("Error Geométrico: La Apotema Inclinada debe ser estrictamente mayor que la mitad del Lado Base para formar una pirámide válida.")
                        return

                    altura_calculada = geometria.calcular_altura_piramide(lado_base, apotema_inclinada_input)
                    widgets_entrada["altura_piramide_input"].value = round(altura_calculada, 2)
                    volumen = geometria.calcular_volumen_piramide(lado_base, apotema_inclinada_input)
                    superficie = geometria.calcular_superficie_piramide(lado_base, apotema_inclinada_input)

                else:
# Ninguna de las entradas requeridas (altura o apotema_inclinada) fue válida o proporcionada.
                    with salida_mensajes:
                         print("Error: Para calcular la Pirámide, debe ingresar un valor positivo para Lado Base y al menos para Altura o Apotema Inclinada.")

            else:
# Lado Base no es válido
                 with salida_mensajes:
                      print("Error: El valor de Lado Base debe ser un número positivo.")


# Actualizar etiquetas de salida si los cálculos fueron exitosos
            if volumen is not None and superficie is not None:
                 widgets_salida["volumen"].value = f"Volumen (cm³): {volumen:.2f}"
                 widgets_salida["superficie"].value = f"Superficie (cm²): {superficie:.2f}"


# Manejo de Errores
# Capturar ValueError, específicamente para math.sqrt() en números negativos (más probable con el método Apotema Inclinada)
    except ValueError as ve:
         with salida_mensajes:
             print(f"Error de Cálculo: Ocurrió un error matemático. Verifique los datos: {ve}")
# Capturar cualquier otra excepción inesperada durante el cálculo
    except Exception as e:
        with salida_mensajes:
            print(f"Ocurrió un error inesperado: {e}")


def al_hacer_click_limpiar(b):
    """
    Manejador de eventos para cuando se hace clic en el botón 'Limpiar'.
    Reinicia los valores de todos los widgets de entrada a 0.0 y borra los resultados y mensajes.

    Args:
        b (Button): El widget de botón que fue clickeado (no se usa directamente en la lógica de la función).
    """
    for key in widgets_entrada:
        widgets_entrada[key].value = 0.0
    widgets_salida["volumen"].value = "Volumen (cm³): --"
    widgets_salida["superficie"].value = "Superficie (cm²): --"
    with salida_mensajes:
        salida_mensajes.clear_output()


# CONFIGURACIÓN Y DISPLAY DE LA INTERFAZ

# Vincular manejadores de eventos a los widgets
selector_figura.observe(al_cambiar_figura, names='value')
boton_calcular.on_click(al_hacer_click_boton)
boton_limpiar.on_click(al_hacer_click_limpiar)

# Configuración inicial: Llamar a actualizar_visibilidad_entrada una vez para establecer las entradas correctas para la figura por defecto
actualizar_visibilidad_entrada(selector_figura.value)

# Organizar los widgets de entrada en una caja vertical.
# Incluir todos los posibles widgets de entrada en el VBox, su visibilidad se controla dinámicamente.
caja_widgets_entrada = VBox([
    widgets_entrada["radio"],
    widgets_entrada["altura"],
    widgets_entrada["lado_base"],
    widgets_entrada["altura_piramide_input"],
    widgets_entrada["apotema_inclinada"],
])

# Organizar los widgets de salida en una caja vertical
caja_widgets_salida = VBox(list(widgets_salida.values()))

# Organizar los botones en una caja horizontal
caja_botones = HBox([boton_calcular, boton_limpiar])

# Combinar todos los componentes principales en un diseño vertical final para la aplicación
disposicion_app = VBox([
    selector_figura,
    caja_widgets_entrada,
    caja_botones,
    caja_widgets_salida,
    salida_mensajes
])

# Mostrar la disposición completa de la aplicación en la salida del notebook
display(disposicion_app)
```
